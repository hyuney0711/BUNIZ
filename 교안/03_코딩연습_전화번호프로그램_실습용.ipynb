# 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

```python
import json

def load_contacts():
    try:
        with open('contacts.json', 'r') as f:
            contacts = json.load(f)
    except FileNotFoundError:
        contacts = {}
    return contacts

def save_contacts(contacts):
    with open('contacts.json', 'w') as f:
        json.dump(contacts, f, indent=2)

def add_contact(contacts, name, number):
    contacts[name] = number
    print(f'{name}의 전화번호가 추가되었습니다.')

def view_all_contacts(contacts):
    print('전체 연락처:')
    for name, number in contacts.items():
        print(f'{name}: {number}')

def search_contact(contacts, name):
    if name in contacts:
        print(f'{name}의 전화번호: {contacts[name]}')
    else:
        print(f'{name}의 전화번호를 찾을 수 없습니다.')

def update_contact(contacts, name, new_number):
    if name in contacts:
        contacts[name] = new_number
        print(f'{name}의 전화번호가 업데이트되었습니다.')
    else:
        print(f'{name}의 전화번호를 찾을 수 없습니다.')

def delete_contact(contacts, name):
    if name in contacts:
        del contacts[name]
        print(f'{name}의 전화번호가 삭제되었습니다.')
    else:
        print(f'{name}의 전화번호를 찾을 수 없습니다.')

def main():
    contacts = load_contacts()

    while True:
        print("\n1. 연락처 추가\n2. 연락처 전체 보기\n3. 검색\n4. 수정\n5. 삭제\n6. 프로그램 종료")
        choice = input('원하는 기능을 선택하세요: ')

        if choice == '1':
            name = input('이름을 입력하세요: ')
            number = input('전화번호를 입력하세요: ')
            add_contact(contacts, name, number)
        elif choice == '2':
            view_all_contacts(contacts)
        elif choice == '3':
            name = input('검색할 이름을 입력하세요: ')
            search_contact(contacts, name)
        elif choice == '4':
            name = input('수정할 이름을 입력하세요: ')
            new_number = input('새로운 전화번호를 입력하세요: ')
            update_contact(contacts, name, new_number)
        elif choice == '5':
            name = input('삭제할 이름을 입력하세요: ')
            delete_contact(contacts, name)
        elif choice == '6':
            save_contacts(contacts)
            print('프로그램을 종료합니다.')
            break
        else:
            print('올바른 선택이 아닙니다. 다시 선택해주세요.')

if __name__ == '__main__':
    main()
```
